#Data Cleaning


```python
df.isnull().sum()
```

```python
df.dropna() # na stands for null value. any row that has null value has been remove
```

```python
df.dropna(how = "any") #any row that has any null value was drop
```

```python
df.dropna(how = "all") # if all the values in any row are null than we drop that row. so here we see that there are not any row has all not value so we can't drop it any row
```

```python
# for fill the missing or null value
df.fillna(df["Age"].mean())  # I am just going to replace all missing value in the age column with the mean of the age.

```

```python
df["Age"].fillna(method = "ffill") # It will be fill the missing values from the forward wise
```

```python
df["Age"].fillna(method = "bfill") # same as it is perform backword
```

```python
df
```

```python
df["Name"]=df["Name"].replace("Charlie","Rose")
```

#Joins and Merges


```python
department_info = {
    "Department": ["HR", "IT", "Finance"],
    "Location": ["New York", "San Francisco", "Chicago"],
    "Manager": ["Laura", "Bob", "Charlie"]
}
```

```python
df2 = pd.DataFrame(department_info)
df2
```

```python
#Concat
pd.concat([df,df2])
```

```python
pd.concat([df,df2],axis = 1) # axis =1 means Full column is drop and if axis = 0 which means 1 row is drop.
```

```python
pd.merge(df, df2, left_on = "Dept", right_on = "Department")
```
